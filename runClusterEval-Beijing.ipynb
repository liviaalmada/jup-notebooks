```python
curs = conn.cursor()
curs.execute("select id as edge_id,km as km from roads")
```

```python
edges = pd.DataFrame(curs.fetchall(), columns=['edge','km'])
```

```python
mapmatching.insert(loc=8, column='speed', value=0.0)
mapmatching.insert(loc=9, column='edge_size', value=0.0)
# tid, edge_id, offset, source_id, target_id, time, x, y
```

```python
trajectories = set(mapmatching.get(0))
```

```python
grouped = mapmatching.groupby(mapmatching[0])
```

```python
print(mapmatching[mapmatching[0]==3])
```

```python
for trajid in list(trajectories):
    trajectory = mapmatching[mapmatching[0]==trajid]
    previous = 0
    
    for actual in np.arange(1,trajectory.shape[0]):
        print(actual)
        edgeid = trajectory.iloc[actual,1]
        size = edges[edges['edge']==edgeid]['km'].get_values()[0]*1000
        time = (trajectory.iloc[actual,5] - trajectory.iloc[previous,5])/1000# de milisegundos para segundos
        speed = size/time     
        previous = actual
        print("speed ", speed)
        print("time ", time)
        print("tamanho ", size)
    break
        
```
