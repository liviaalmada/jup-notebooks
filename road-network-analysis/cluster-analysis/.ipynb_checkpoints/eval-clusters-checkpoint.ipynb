```python
import pandas as pd
from scipy import stats as st
```

```python
data = pd.read_csv('clusters-ex', delimiter=';', header=None)
```

```python
data.head()
```

```python
d = data[data[4]>=0]
```

```python
d = d[d[5]>=0]
d.head()
```

```python
d1 = d[d[0]==4]
d1 = d1.get_values()
d1.size
```

```python
for i in range(0,len(d1)):
    for j in range(0,len(d1)):
        if(i!=j):        
            x = st.kstest(st.norm.rvs(size=100, loc=d1[i,4], scale=d1[i,5]),"norm", args=(d1[j,4],d1[j,5]))
            print(i, j, x.statistic)
            
```

```python
def eval_cluster(cluster):
    n = 0
    for i in range(0,len(cluster)):
        for j in range(0,len(cluster)):
            if(i!=j):        
                x += st.kstest(st.norm.rvs(size=100, loc=cluster[i,4], scale=cluster[i,5]),"norm", args=(cluster[j,4],cluster[j,5]))
                n++
    return x/n
```
