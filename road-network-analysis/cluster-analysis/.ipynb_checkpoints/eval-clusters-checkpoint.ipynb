```python
def print_hist_size(filepath):
    clusters, noise = eval.read_cluster_file(filepath)
    sizes = eval.clusters_size(clusters)
    %config InlineBackend.figure_format = 'pdf'
    plt.ylabel('Number of Clusters')
    plt.xlabel('Number of Links')
    plt.title('Size of Clusters')
    plt.hist(sizes[1:])
    plt.show()    
```

### Print histograms of clusters size

```python
print_hist_size('/Users/liviaalmada/NYDataset/Link Level Traffic Estimates/link-5-10-15-10')
print_hist_size('/Users/liviaalmada/NYDataset/Link Level Traffic Estimates/link-4-10-15-10')
print_hist_size('/Users/liviaalmada/NYDataset/Link Level Traffic Estimates/link-3-10-15-10')
print_hist_size('/Users/liviaalmada/NYDataset/Link Level Traffic Estimates/link-2-10-15-10')
print_hist_size('/Users/liviaalmada/NYDataset/Link Level Traffic Estimates/link-1-10-15-10')
```
